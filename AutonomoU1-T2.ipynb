## 1.2.1 Herencia y Sintaxis de Herencia

Primero se crea una clase base llamada ModeloIA.  
Esta clase tiene métodos generales para entrenar, predecir y evaluar.

Luego se crean dos clases hijas: ClasificadorIA y RegresorIA.  
Cada una hereda de ModeloIA, pero cambia algunos métodos para simular comportamientos diferentes.

```python
class ModeloIA:
    def __init__(self, nombre, precision):
        self.nombre = nombre
        self.precision = precision

    def entrenar(self, datos):
        # Simula el entrenamiento de un modelo general
        print("Entrenando el modelo:", self.nombre)
        print("Datos usados:", datos)

    def predecir(self, entrada):
        # Predicción general
        print("El modelo", self.nombre, "hace una predicción con:", entrada)
        return "Predicción general"

    def evaluar(self, resultados_reales, resultados_predichos):
        # Evaluación simulada
        print("Evaluando el modelo:", self.nombre)
        print("Resultados reales:", resultados_reales)
        print("Resultados predichos:", resultados_predichos)
        print("Precisión simulada:", self.precision)

    def __str__(self):
        # Devuelve una descripción entendible del objeto
        return "Modelo IA: " + self.nombre + " con precisión de " + str(self.precision)

    def __add__(self, otro_modelo):
        # Fusiona dos modelos creando uno nuevo
        nuevo_nombre = self.nombre + " + " + otro_modelo.nombre
        nueva_precision = (self.precision + otro_modelo.precision) / 2
        return ModeloIA(nuevo_nombre, nueva_precision)

    def __sub__(self, otro_modelo):
        # Calcula la diferencia de precisión entre dos modelos
        diferencia = self.precision - otro_modelo.precision
        return diferencia


class ClasificadorIA(ModeloIA):
    def entrenar(self, datos):
        # Se reutiliza parte del método de la clase padre
        super().entrenar(datos)
        print("Entrenamiento especial para clasificar datos.")

    def predecir(self, entrada):
        # Simula una clasificación
        print("El clasificador analiza la entrada:", entrada)
        return "Clase A"


class RegresorIA(ModeloIA):
    def entrenar(self, datos):
        # Se reutiliza parte del método de la clase padre
        super().entrenar(datos)
        print("Entrenamiento especial para calcular valores numéricos.")

    def predecir(self, entrada):
        # Simula una regresión
        print("El regresor analiza la entrada:", entrada)
        return 75.5
```

## 1.2.2 Polimorfismo y uso de super()

Aquí se crean objetos de diferentes clases.  
Aunque todos usan el método predecir(), cada objeto responde de forma distinta según su clase.

Esto demuestra el polimorfismo.

```python
modelo_general = ModeloIA("Modelo Básico", 70)
clasificador = ClasificadorIA("Clasificador de Imágenes", 85)
regresor = RegresorIA("Regresor de Ventas", 80)

print("=== Entrenamiento de modelos ===")
clasificador.entrenar(["foto gato", "foto perro", "foto carro"])
print()
regresor.entrenar([100, 200, 300, 400])

print("\n=== Polimorfismo ===")

modelos = [modelo_general, clasificador, regresor]

for modelo in modelos:
    print("\nUsando:", modelo.nombre)
    respuesta = modelo.predecir("dato de prueba")
    print("Resultado:", respuesta)
```

## 1.2.3 Métodos mágicos y sobrecarga de operadores

En esta parte se usan métodos mágicos.

El método __str__ permite imprimir el objeto de forma más clara.  
El método __add__ permite sumar o fusionar dos modelos.  
El método __sub__ permite restar la precisión simulada entre dos modelos.

```python
print("=== Método __str__ ===")
print(clasificador)
print(regresor)

print("\n=== Método __add__ ===")
modelo_fusionado = clasificador + regresor
print(modelo_fusionado)

print("\n=== Método __sub__ ===")
diferencia = clasificador - regresor
print("Diferencia de precisión entre los modelos:", diferencia)
```

## 1.2.4 Patrón Singleton y Gestión de Sesiones

En esta parte se crea una clase GestorSesion.  
El patrón Singleton sirve para que solo exista una instancia del gestor.

Esto puede servir en sistemas de IA para controlar una sola conexión o una sola sesión de entrenamiento.

```python
class GestorSesion:
    _instancia = None

    def __new__(cls):
        # Si no existe una instancia, se crea una
        if cls._instancia is None:
            cls._instancia = super(GestorSesion, cls).__new__(cls)
        return cls._instancia

    def iniciar_sesion(self):
        # Simula conexión a un servidor de IA
        print("Sesión iniciada con el servidor de IA.")

    def cerrar_sesion(self):
        # Simula cerrar conexión
        print("Sesión cerrada correctamente.")


sesion1 = GestorSesion()
sesion2 = GestorSesion()
sesion3 = GestorSesion()

sesion1.iniciar_sesion()

print("ID sesión 1:", id(sesion1))
print("ID sesión 2:", id(sesion2))
print("ID sesión 3:", id(sesion3))

if sesion1 is sesion2 and sesion2 is sesion3:
    print("Todas las sesiones son el mismo objeto.")
else:
    print("Las sesiones son diferentes.")

sesion1.cerrar_sesion()
```

## 1.2.5 Reflexión Final

La herencia ayuda bastante porque no toca volver a escribir el mismo código muchas veces. Por ejemplo aquí primero se hizo una clase general y luego las otras clases heredaron varias cosas de ahí. Eso hace que el código sea más organizado y más fácil de entender.

El polimorfismo me pareció interesante porque aunque todos usan el método predecir(), cada modelo responde diferente dependiendo del tipo de modelo que sea. Entonces Python reconoce automáticamente qué método usar.

Los métodos mágicos ayudan porque hacen que los objetos tengan funciones más prácticas. Por ejemplo con __str__ se puede imprimir mejor la información del modelo y con __add__ o __sub__ se pueden hacer operaciones entre modelos de una forma más sencilla.

El patrón Singleton sirve para que no se creen muchas sesiones al mismo tiempo. En sistemas de IA eso puede ayudar para controlar recursos o conexiones y evitar problemas si varias sesiones intentan hacer lo mismo.
